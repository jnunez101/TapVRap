```python
import csv

with open("set4Abutton.txt", "r") as data:
    mylist = [row[:] for row in csv.reader(data, delimiter=',')]
print(mylist)
```

```python
from string import ascii_letters, punctuation, whitespace
import re
nono = ['(', ')']
newlist = []
for i in range(len(mylist)):
    if len(mylist[i]) < 5:
        continue
    for y in range(len(mylist[i])):
        mylist[i][y] = re.sub(r'[^\d.]+', '', mylist[i][y])
        mylist[i][y] = "".join(filter(lambda x: x not in nono, mylist[i][y]))
        mylist[i][y] = float(mylist[i][y])
        #print(mylist[i][y])
```

```python
for i in range(len(mylist)):
    if type(mylist[i][0]) is str:
        continue
    else:
        newlist.append(mylist[i])
```

```python
import matplotlib as plt

def Sort(mylist):
    return(sorted(mylist, key = lambda x: x[1]))

print(Sort(newlist))
```

```python
import matplotlib.pyplot as plt
import matplotlib
finger = 0
time = []
z = []
for i in range(len(newlist)):
    if newlist[i][1] == finger:
        time.append(newlist[i][0])
        z.append(newlist[i][4])
fig, ax = plt.subplots()
ax.plot(time,z)
ax.set(xlabel='time', ylabel='Z', title='uwu')
ax.grid()
plt.show()
print(max(z))
print(min(z))
#ax.plot(newlist[y,0], newlist[y,4])

```

```python
#time = []
x = []
for i in range(len(newlist)):
    if newlist[i][1] == finger:
        #time.append(newlist[i][0])
        x.append(newlist[i][2])
fig, ax = plt.subplots()
ax.plot(time,x)
ax.set(xlabel='time', ylabel='X', title='uwu')
ax.grid()
plt.show()
print(max(x))
print(min(x))
```

```python
y = []
for i in range(len(newlist)):
    if newlist[i][1] == finger:
        #time.append(newlist[i][0])
        y.append(newlist[i][3])
fig, ax = plt.subplots()
ax.plot(time,y)
ax.set(xlabel='time', ylabel='Y', title='uwu')
ax.grid()
plt.show()
print(max(y))
print(min(y))
```
